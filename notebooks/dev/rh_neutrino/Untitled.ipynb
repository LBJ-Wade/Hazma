```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from hazma.rh_neutrino import RHNeutrino
```

```python
model = RHNeutrino(300.0, 1e-3, lepton='e', include_3body=True)
```

```python
def state_to_tex(state):
    if state == 'pi l':
        return r'$\pi^{\pm} + \ell^{\mp}$'
    elif state == 'pi0 nu':
        return r'$\pi^{0} + \nu$'
    elif state == 'k l':
        return r'$K^{\pm} + \ell^{\mp}$'
    elif state == 'nu l l':
        return r'$\nu + \ell^{\mp} + \ell^{\mp}$'
    elif state == 'l pi pi0': 
        return r'$\pi^{\pm} + \ell^{\mp} + \pi^{0}$'
    elif state == 'nu pi pi': 
        return r'$\nu + \pi^{\pm} + \pi^{\mp}$'
    elif state == 'nu g g':
        return r'$\nu + \gamma + \gamma$'
    elif state == 'total':
        return r'$\mathrm{Total}$'
    else:
        return state
```

```python
engs = np.logspace(1e-3, np.log10(model.mx))
spectra = model.spectra(engs)

plt.figure(dpi=150)

for state, dnde in spectra.items():
    plt.plot(engs, dnde, label=state_to_tex(state))
plt.yscale('log')
plt.xscale('log')
plt.ylabel(r'$\frac{dN}{dE_{\gamma}}$', fontsize=16)
plt.xlabel(r'$E_{\gamma}$', fontsize=16)
plt.legend()
```
